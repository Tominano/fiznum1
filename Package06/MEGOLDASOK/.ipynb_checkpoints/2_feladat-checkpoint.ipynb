## 02 - Kettős integrálok

- Határozzuk meg az $$y=\pi^2/4-(x-\pi/2)^2)$$ és $$y=-sin(3x))$$ görbék által határolt tartomány területét.
- Integráljuk az fenti tartományon az $$f(y,x)=e^{-(x-\pi/2)^2-(y-\pi/2)^2} $$ függvényt.
- Ábrázoljuk az integrálási tartományt, illetve az integrálandó $f(y,x)$ függvényt egy ábrán!


```python
# a szokásos rutinok betöltése
%pylab inline
from scipy.integrate import * # az integráló rutinok betöltése
```

```python
def f(x,y):
    return exp(-(x-pi/2)**2)*exp(-(y-pi/2)**2)  #az egyik intgrálandó függvény megadása

dblquad(f, 0, pi, lambda x:(-sin(3*x)), lambda x:pi**2/4-(x-pi/2)**2)  #függvénydeklarációk megadása

#dblquad(f, -1/2, 1/2, lambda x:-1/2, lambda x:1/2)
#dblquad(g,0,pi, lambda x:0, lambda x:pi)
```

```python
x2,y2 = meshgrid(linspace(0,pi,100),linspace(-1,3,100))  #a mintavételezés legyártása az f függvényhez
pcolor(x2,y2,f(x2,y2))   #a pcolor ábrázolás
x=linspace(0,pi)  #mintavételezés legyártása a t,z hez
t=pi**2/4-(x-pi/2)**2
z=(-sin(3*x))    
plot(x,t)
plot(x,z)   #majd egy ábrán ábrázoljuk a megoldásokat

title('A három függvény')   #tengelyeliratokat és nevet adok az ábrának
ylabel('Y')
xlabel('X')
```

```python
x=linspace(0,pi)
def t(x):
    return pi**2/4-(x-pi/2)**2  #definiálom a függvényeket
def z(x):
    return -sin(3*x)
```

```python
te=quad(t,0,pi)
ze=quad(z,0,pi)  #integgrálom a függvényeket amikből a terület számolható

print(te,ze)
```
