### Interaktív ábrák II
Készíts egy interaktív ábrát mely egy egység sugarú kört ábrázol melybe egy szabályos $N$ csúcsú sokszög van rajzolva. $N$ értékét egy csúszka segítségével lehessen változtatni. A kör közepébe írjuk bele a sokszög kerületének a hosszának felét!


```python
%pylab inline
from ipywidgets import *  # az interaktivitásért felelős csomag

figsize(6,6)
```

```python
def n(sides, radius=1, rotation=0, translation=None):
    sides = sides+1
    one_segment = math.pi * 2 / sides
    
    points = [
        (math.sin(one_segment * i + rotation) * radius,
         math.cos(one_segment * i + rotation) * radius)
        for i in range(sides)]
    
    points.append((0.0, 1.0))
    
    if translation:
        points = [[sum(pair) for pair in zip(point, translation)]
                  for point in points]

    return points
```

```python
def abra(N,szin):
    phi=arange(0,2*pi,0.01)
    r=1
    a = n(N-1)
    l=N*sqrt((a[0][0]-a[1][0])**2+(a[0][1]-a[1][1])**2)/2
    
    plot(r*cos(phi), r*sin(phi),color=szin)
    plot(*zip(*a),color=szin)
    annotate('%f'%(l,), xy=(-1, 0), xytext=(-0.3, 0))
    xlim(-1.5,1.5)
    ylim(-1.5,1.5)
    grid(True)
interact(abra,N=(3,18),szin=('blue','green','black','yellow','red'));
```
