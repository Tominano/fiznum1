### Egyszerű plot

Írj egy függvényt, ami kiszámolja a következő értéket egy bemeneti `x` értékből: $$\frac{x^2}{e^x \cdot tan(x)}$$


- Ábrázold -2 és 2 között 50 pontban (egyenletes mintavételezéssel) a fent leírt függvényt. 
- Az ábrádnak legyen címe és legyenek tengelyfeliratai (ezeket te választod meg). 
- Az adatokat "mediumaquamarine" színű telt csillagokkal ábrázold. 
- A csillagok mérete legyen 12-es. 
- Az ábrán kék nyíllal jelöld meg a lokális minimum helyét ("lokális minimum" felirattal, 20 pontos szöveggel). A lokális minimum helyét elég a kiszámított numerikus értékek alapján meghatározni.

A kapott ábrát mentsd ki "plot.jpg" néven, majd egy ezt követő markdown cellába töltsd is be!

```python
%pylab inline #meghívom a szükséges csomagot a számításhoz és ábrázoláshoz
```

```python
x=linspace(-2,2,50) #megadom az ábrázolandó pontok számát és intervallumát
y=(x**2)/(exp(x)*tan(x)) #definiálom az ábrázolandó függvényt
```

```python
ym=min(y) #meghatározom a minimum értéket
xm=x[ym==y] #meghatározom az érték helyét 
```

```python
plot(x,y,'*', color='mediumaquamarine', markersize=12) #ábrázolom a függvényt aés meghatározom az ábrázolás formáit
xlabel('tartomány(x)') # elnevezem a tengelyeket
ylabel('értékek(y)')
annotate('Lokális minimum', xy=(xm, ym+0.5), xytext=(-1,3),arrowprops=dict(color='blue',width=3),fontsize=20)
# a lokális minimumot meghatározom és megjelölöm a helyét a már definiált függvénnyel, és megformázom
savefig("plot.jpg") #kimentem a képet
```
