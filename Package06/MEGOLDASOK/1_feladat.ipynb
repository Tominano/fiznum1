## 01 - Egyszerű integrálás

A quad függvény segítségével integráljuk a
$$ f(x)=\cos^2(4 \arccos(x)) $$
függvényt az [0,1] intervallumon!

```python
# a szokásos rutinok betöltése
%pylab inline
from scipy.integrate import * # az integráló rutinok betöltése
```

```python
def f(x):
    return cos(4*arccos(x))**2 #definiálom a függvényemet
```

```python
quad(f,0,1)  # majd kiintegrálom a függvényt a határokon
```
