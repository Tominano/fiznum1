# 5. feladat

Adott a **data/hf.hdf5** állomány. A tanult módon olvasd be a pythonba, majd végezd el a következő feladatokat:

- Keresd meg benne a csoportok neveit!
- Keresd meg azt az adattömböt, melynek dimenziói: 64x64x64
- Ábrázold ennek a tömbnek a 3. koordináta szerinti 32.-dik szeletét
- Az ábra az `imshow` használatával, "viridis" színséma beállítással készüljön.

```python
import h5py #importálom a megfelelő csomagokat
```

```python
%pylab inline
```

```python
f = h5py.File("../data/hf.hdf5", "r") #definiálom ugy hogy beolvasom úgy hogy olvasni tudjam

list(f.keys()) # ha nem rendezzük listába akkor nem írja ki szépen!
```

```python
list(f["xxlprho_128"].keys())  # alsóbb rétegekben is keresünk
```

```python
f["xxlprho_128/value"] # # Ebben van a keresett adat tömb
```

```python
f["xxlprho_128/value"][:]  #arrayt készítünk az adatokból
```

```python
imshow(f["xxlprho_128/value"][:,:,32],cmap="viridis") #ábrázolom a 32.szeletet a 3-vektor szerint. Köszi!
```
