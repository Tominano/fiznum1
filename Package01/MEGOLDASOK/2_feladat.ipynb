## 2. Feladat

#### Egyszerű ábrázolás és Függvények - I.


Ebben a feladatban adott két lista, amely egy műveleti erősítős integráló RC-kör frekvenciafüggő átviteli erősítési adatit tartalmazza, a frekvenciaadatokat $\mathrm{Hz}$-ben, az átviteli arányt $\mathrm{V}/\mathrm{V}$-ban.

```python
%pylab inline

frekvencia=[50.23, 100, 200.5, 499.5, 1000, 1593, 2007, 5017, 10040, 20000, 49800]
atvitel=[30,15.55, 8.8889, 3.1136, 1.5454, 0.9773, 0.7674, 0.2954, 0.2272, 0.1704, 0.1363]

atvitel_db=[10*math.log(x**2,10) for x in atvitel] #az átviteli skálát decibelre váltom

def pol(x1,x2,y1,y2):
    X = (x1+x2)/2
    Y = y1+(((X-x1)*(y2-y1))/(x2-x1)) #létrehozom a függvényt a képlet alapján
    return X, Y


X1 = frekvencia  #elnevezem az ábrázoláshoz az adatokat
Y1 = atvitel_db
Y = []
X = []

for elem in range(len(frekvencia)-1): #Elkészítem az x,y változókat a függvénnyel
    XY = pol(frekvencia[elem], frekvencia[elem+1], atvitel_db[elem], atvitel_db[elem+1])
    X.append(XY[0])
    Y.append(XY[1])



xlabel("Frekvencia [Hz]") #majd ábrázolom az adatokat
ylabel("Átvvitali arány [dB]")
title("Adatok lineáris interpolációval")
plot(X1,Y1, '.')
plot(X,Y, 'g+')
```
