# *Az [Abramowitz and Stegun](http://people.math.sfu.ca/~cbm/aands/) könyvből az alábbi képletet választottam:*

### 12.2.2-es egyenlet


$$\mathrm{e}^{-\frac{1}{2}z^2}\cdot M \left(-\frac{1}{2}a+\frac{1}{4},\frac{1}{2},-\frac{1}{2}x^2 \right)$$
$$=\mathrm{e}^{-\frac{1}{2}z^2} \left\{1+ \left(a-\frac{1}{2} \right)\cdot\frac{x^2}{2!}+ \left(a-\frac{1}{2} \right)\cdot \left(a-\frac{5}{2} \right)\cdot\frac{x^4}{4!}+... \right\}$$
