# 3. feladat

### 3D-s ábrák - I.

Ábrázoljuk az alábbi, paraméteres alakban megadott térbeli görbét:

\begin{align} 
x&=\sin(t) + 2 \sin (2t),\\
y&=\cos(t) - 2 \cos (2t),\\
z&=-\sin(3t).
\end{align}

- A $t$ paramétert $t\in[0,2\pi]$ intervallumban vegyük fel! Készítsünk képet két különböző nézőpontból!
- Mindkét ábrán szaggatott vonallal rajzolj, és a vonalvastagság legyen 3-as.
- Az ábráknak legyenek címei és tengelyfeliratai!

```python
%pylab inline      #az ábrázoláshoz és számításokhoz meghívott csomag
from mpl_toolkits.mplot3d import * #3D-s ábrák alcsomagja
```

```python
t=linspace(0,2*pi,100) # 100 pont 0 és 2*pi között
ax=subplot(1,1,1,projection='3d') #térbeli koordináta tengely létrehozása
ax.plot(sin(t)+2*sin(2*t),cos(t)-2*cos(2*t),-sin(3*t),linestyle='dashed',linewidth=3)#ábrázolom a 3d függvényt, beállítom a szaggatott 3-as vonalat.
figsize(12,12) #méretet állítok
ax.view_init(0,0) # néző szöget állítok
title('Ábra') #nevet adok az ábrának
ax.set_xlabel('x tengely',labelpad=1) 
ax.set_ylabel('y tengely',labelpad=1)  #elnevezem a tengelyeket is
ax.set_zlabel('z tengely',labelpad=1)

```

```python
t=linspace(0,2*pi,100) # 100 pont 0 és 2*pi között
ax=subplot(1,1,1,projection='3d') #térbeli koordináta tengely létrehozása
ax.plot(sin(t)+2*sin(2*t),cos(t)-2*cos(2*t),-sin(3*t),linestyle='dashed',linewidth=3)
ax.view_init(0,90)  #egyedül ez változik és látható hogy más szögből mutatja az ábrát
figsize(12,12)
title('Ábra')
ax.set_xlabel('x tengely',labelpad=1)
ax.set_ylabel('y tengely',labelpad=1)
ax.set_zlabel('z tengely',labelpad=1)
```

```python
#itt nem állítom be hogy honnan nézem ezért látszik egyenlő szög alatt mind a három oldal
t=linspace(0,2*pi,100) # 100 pont 0 és 2*pi között
ax=subplot(1,1,1,projection='3d') #térbeli koordináta tengely létrehozása
ax.plot(sin(t)+2*sin(2*t),cos(t)-2*cos(2*t),-sin(3*t),linestyle='dashed',linewidth=3)
figsize(12,12)
title('Ábra')
ax.set_xlabel('x tengely',labelpad=1)
ax.set_ylabel('y tengely',labelpad=1)
ax.set_zlabel('z tengely',labelpad=1)
```
