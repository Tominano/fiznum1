### Magic - Fortran-Python

Python kernel fortran magic-jét használva definiáljunk egy fortran függvényt a mintapéldában látottak alapján ami két bemenő vektor átlagát számolja ki. 

```python
%lsmagic
%pylab inline
%load_ext rpy2.ipython 
%load_ext fortranmagic
```

```python
%%fortran
       subroutine fort_funct(u, v, w)
            real, intent(in) :: u(:), v(:)  !ezek bemeno valtozok
            real, intent(out) :: w(size(u)) !ez lesz a kimeno valtozo
            !vektor muveleteket használunk
            w(:) = (u(:) + v(:))/2          !a W a két vektor átlaga
       end subroutine
```

## Majd lepróbálom a függvényemet

```python
u=[50,45]
v=[100,56]   # definiálok két vektort
```

```python
fort_funct(u,v) # majd lepróbálom a függvényemet
```
