## 05 Hol vannak a fekete lyukak?

A `../data/black_hole.npy` binárisan tárolt file-ban a [LIGO](https://www.ligo.caltech.edu/) detektor legfrissebb mérési adatai találhatóak. Az adatsort a detektor 44100 Hz-es mintavételezési frekvenciával vette fel. 
- Ábrázold az adatsor jelalakját !
- Határozd meg a jel Fourier-spektrumának legnagyobb és legkissebb frekvenciáját.
- A jel spektrogramjának segítségével becsüld meg a detektált fekete lyuk(ak) koordinátáit.

```python
%pylab inline
from scipy.optimize import curve_fit # Az illesztéshez használt függvény
from numpy.fft import *              # Fourier-analízishez használt rutinok
from scipy.signal import spectrogram # Spektrogramm készítő függvény
```

```python
y=load('../data/black_hole.npy') # beolvasom 
x=linspace(0,len(y)/44100,len(y)) # legyártom a pontokat
plot(x,y) # plotolom
print(max(y), min(y)) # a max és a min kiiratása
```

```python
Fjel=fft(y) # Fourier-transzformáció
freq = fftfreq(len(x),x[1]-x[0]) # frekvenciia adatok
plot(freq,abs(Fjel)) # a Fourier nagysága
```

```python
# csak a pozitívokat nézem mert időnél elegendő azzal dolgozni
figsize(16,4)
plot(freq,abs(Fjel))
xlim(0,23000)
ylim(0,16000000)
```
