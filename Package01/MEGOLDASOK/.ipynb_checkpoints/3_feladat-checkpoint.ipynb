#### Dictionary - I.

Írjatok egy függvényt, ami a megadott szótár segítségével tetszőleges magyar szöveget angolra fordít!

- A fordítás történjen szóról szóra.
- Ha ismeretlen szóval találkoztok, akkor csináljátok azt, amit a Google Translator: csak írjátok ki az eredeti szót fordítás nélkül. ( A szöveg többi részét természetesen le kell fordítani.)
- Nem baj, ha nyelvtanilag nem helyes vagy értelmetlen a fordítás.

```python
pelda_szoveg='a majom szereti a banánt minden kutya szereti a kolbászt'

# Létrehozom a traslate functiont, ami szét darabolja a beírt szöveget és kicseréli a szótárban lévő szavakra, ha létezik.

def translate(szó):
    transdict={'a':'the',
        'majom': 'monkey',
        'kutya':'dog',
        'szereti':'likes',
        'banánt':'banana',
        'kolbászt':'sausage'}
    words = szó.split()
    trans = [transdict.get(w.lower(),w) for w in words]
    print(' '.join(trans))
    

def userinput():
    szó = input("Milyen mondatot szeretnél lefordítani? ") #ezzel pedig definiáljuk az inputot
    translate(szó)

userinput() #meghívom az előbb definiált függvényt
```
